# Scraping

```python
import json
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# Setup WebDriver (automatically downloads the appropriate ChromeDriver)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# List of URLs you want to scrape
adv_urls = ["https://www.swiggy.com"]


# Function to scrape each URL
def scrape_website(url):
    # Open the website and wait for it to load
    driver.get(url)
    driver.implicitly_wait(10)
    # driver.set_page_load_timeout(30)  # Adjust wait time if necessary

    # Get the rendered page source
    page_source = driver.page_source

    # Parse the page source with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # ----------------- Data Extraction -----------------
    # Extract title and meta description (if available)
    title = soup.title.get_text(strip=True) if soup.title else ""
    meta_desc = ""
    meta_tag = soup.find("meta", attrs={"name": "description"})
    if meta_tag and meta_tag.get("content"):
        meta_desc = meta_tag.get("content")

    # Extract headings: h1, h2, h3 (these might capture key info about the merchant)
    headings = {
        "h1": [tag.get_text(strip=True) for tag in soup.find_all("h1")],
        "h2": [tag.get_text(strip=True) for tag in soup.find_all("h2")],
        "h3": [tag.get_text(strip=True) for tag in soup.find_all("h3")],
    }

    # Extract paragraphs text (often used for product details, descriptions, etc.)
    paragraphs = [tag.get_text(strip=True) for tag in soup.find_all("p")]

    # Optionally, also extract any list items if they are used for products/services
    list_items = [tag.get_text(strip=True) for tag in soup.find_all("li")]

    # Combine all text content (this "full_text" can be useful for context)
    full_text = " ".join([title, meta_desc] + headings["h1"] + headings["h2"] + headings["h3"] + paragraphs + list_items)

    # ----------------- Structured Output -----------------
    return {
        "title": title,
        "meta_description": meta_desc,
        "headings_h1": ", ".join(headings["h1"]),
        "headings_h2": ", ".join(headings["h2"]),
        "headings_h3": ", ".join(headings["h3"]),
        "paragraphs": ", ".join(paragraphs),
        "list_items": ", ".join(list_items),
        "full_text": full_text
    }

# ----------------- Scraping Multiple Websites -----------------
all_merchant_data = []
all_adv_data=[]

for url in urls:
    merchant_data = scrape_website(url)
    merchant_data['url'] = url  # Add the URL as an additional field
    all_merchant_data.append(merchant_data)
for adv_url in adv_urls:
    adv_data = scrape_website(adv_url)
    adv_data['url'] = url  # Add the URL as an additional field
    all_adv_data.append(adv_data)


# advertiser_data = scrape_website(advertiser_url)
# advertiser_data['url'] = url  # Add the URL as an additional field
# all_advertiser_data.append(advertiser_data)   

# Convert the list of dictionaries to a pandas DataFrame
df = pd.DataFrame(all_merchant_data)
df_adv = pd.DataFrame(all_adv_data)
# Output the DataFrame to see the result
# print(df)

# Optionally, save the DataFrame to a CSV file
# df.to_csv("merchant_data.csv", index=False)

# Close the browser after scraping
driver.quit()
```
